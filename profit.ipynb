```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (9,6)
```

```python
colors = plt.cm.YlGnBu([.4, .55, .7, .85, 1])
```

```python
profit = lambda a_0, N, A=10000000: (a_0 * N * (a_0 + 2*A + N)) / (a_0**2 + 2*a_0*A + a_0*N + A**2)
```

```python
n_range = np.linspace(0, 100, 2000)

for i, A in enumerate([20, 40, 80, 200, 1000000]):
    p = profit(A/100, n_range, A)
    if i == 0:
        plt.plot(n_range, p, c=colors[i], label='Illiquid market (small constant)')
    elif i == 4:
        plt.plot(n_range, p, c=colors[i], label='Liquid market (large constant)')
    else:
        plt.plot(n_range, p, c=colors[i])
        
plt.xlabel('Number of tokens used to push the exchange rate ($N$)', fontsize=8)
plt.ylabel('Profit from a sandwich ($P$)', fontsize=8)
plt.title("Market Liquidity's Effect on Arbitrage Profit")
plt.legend(fontsize=8)
plt.xlim(0, 100)
plt.savefig('profit_N.png')
plt.show()
```

```python
a_range = np.linspace(0, 100, 2000)

for i, A in enumerate([20, 50, 100, 200, 500]):
    p = profit(a_range, 10, A)
    if i == 0:
        plt.plot(n_range, p, c=colors[i], label='Illiquid market (small constant)')
    elif i == 4:
        plt.plot(n_range, p, c=colors[i], label='Liquid market (large constant)')
    else:
        plt.plot(n_range, p, c=colors[i])
        
plt.xlabel("Number of tokens used for the arbitrageur's initial trade ($a_0$)", fontsize=8)
plt.ylabel('Profit from a sandwich ($P$)', fontsize=8)
plt.title("Market Liquidity's Effect on Arbitrage Profit")
plt.legend(fontsize=8)
plt.xlim(0, 100)
plt.savefig('profit_a.png')
plt.show()
```

```python
a_range = np.linspace(0, 100, 2000)

for i, A in enumerate([20, 50, 100, 200, 500]):
    p = profit(a_range, 10, A)/a_range
    if i == 0:
        plt.plot(n_range, p, c=colors[i], label='Illiquid market (small constant)')
    elif i == 4:
        plt.plot(n_range, p, c=colors[i], label='Liquid market (large constant)')
    else:
        plt.plot(n_range, p, c=colors[i])
        
plt.xlabel("Number of tokens used for the arbitrageur's initial trade ($a_0$)", fontsize=8)
plt.ylabel('Return from a sandwich ($P/a_0$)', fontsize=8)
plt.title("Market Liquidity's Effect on Arbitrage Return")
plt.legend(fontsize=8)
plt.xlim(0, 100)
plt.savefig('return_a.png')
plt.show()
```
