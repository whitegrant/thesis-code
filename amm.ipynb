```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 150
```

```python
color = plt.cm.YlGnBu(.75)
```

```python
n = 3
domain = np.geomspace(1/n, n, 1000)
plt.plot(domain, 1 / domain, c=color)
plt.title('A x B = C')
plt.xlim(0, 3)
plt.xlabel('A')
plt.ylim(0, 3)
plt.ylabel('B')
plt.tick_params(
    axis='both',
    which='both',
    bottom=False,
    left=False,
    labelbottom=False,
    labelleft=False)
plt.gca().set_aspect('equal')
plt.savefig('amm.png')
plt.show()
```
