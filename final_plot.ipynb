```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (9,6)
# plt.rcParams['figure.figsize'] = (6,4)
```

```python
colors = plt.cm.YlGnBu([.6, .75, 1.])
```

### Fix Block Size, Vary Number of Pending Trades

```python
# nL = np.load('results/new_size_naive.npy')
# sL = np.load('results/new_size_single.npy')
# mL = np.load('results/new_size_mev.npy')

nL = np.load('results/size_naive.npy')
sL = np.load('results/size_single.npy')
mL = np.load('results/size_mev.npy')
```

```python
nL2 = np.load('results/size_naive2.npy')
sL2 = np.load('results/size_single2.npy')
mL2 = np.load('results/size_mev2.npy')
```

```python
domain = np.arange(4, 14)*6
```

```python
len(domain)
```

```python
nL = np.vstack((nL, nL2))
sL = np.vstack((sL, sL2))
mL = np.vstack((mL, mL2))
```

```python
nL.shape
```

```python
n_median = np.median(nL, axis=1)
n_min = np.percentile(nL, 25, axis=1)
n_max = np.percentile(nL, 75, axis=1)

s_median = np.median(sL, axis=1)
s_min = np.percentile(sL, 25, axis=1)
s_max = np.percentile(sL, 75, axis=1)

m_median = np.median(mL, axis=1)
m_min = np.percentile(mL, 25, axis=1)
m_max = np.percentile(mL, 75, axis=1)
```

```python
# plt.fill_between(domain, n_min, n_max, color=colors[0], alpha=.3)
# # plt.fill_between(domain, s_min, s_max, color=colors[1], alpha=.3)
# plt.fill_between(domain, m_min, m_max, color=colors[2], alpha=.3)

plt.semilogy(domain, n_median, c=colors[0], label='tips only')
plt.semilogy(domain, s_median, c=colors[1], label='single sandwich')
plt.semilogy(domain, m_median, c=colors[2], label='tips and arbitrage')

plt.title('Profit from Using Sandwiches in Block Creation')
plt.xlim(domain[0], domain[-1])
plt.xlabel('Number of Pending Trades')
plt.ylabel('Profit')
plt.legend()
plt.savefig('test.png')
plt.show()
```
